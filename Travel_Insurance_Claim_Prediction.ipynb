```
from pycaret.classification import *
```

```
s=setup(data=train_df,target='Claim',session_id=102,
        normalize=True,transformation=True,polynomial_features=True,trigonometry_features=True,
        feature_selection=True,fix_imbalance=True,remove_outliers=True,
        remove_multicollinearity=True,use_gpu=True,polynomial_degree=3)
```

```
compare_models(sort='F1')
```

```
!pip install optuna
```

```
lgbm= create_model('lightgbm')
```

```
#tuning lgbm using optuna
tuned_lgbm= tune_model(lgbm,optimize='F1',search_library='optuna')
```

```
"""
Bagging is default here
It  reduces variance and helps to avoid overfitting
"""
ensembled_LGBM= ensemble_model(tuned_lgbm,optimize='F1')      
```

```
evaluate_model(ensembled_LGBM)
```

```
predictions = predict_model(ensembled_LGBM,data=test_df)['Label']
```

```
predictions
```

```
res = pd.DataFrame(predictions)

res.index=test_df.index
res.columns = ['prediction']

res.to_csv('Pycaret-LGBM.csv',index=False)
```
